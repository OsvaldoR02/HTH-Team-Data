```python
import pandas
import seaborn
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipotle_data = pandas.read_csv(url, sep = '\t')

chipotle_data.head()
```

```python
revenue = 0

for index, row in chipotle_data.iterrows():
    # after accessing item price we need to multiply price by quantity and add this total to our revenue total
    
    # print(row["item_price"].replace("$",""))
   # print(row["quantity"])
   # print(row["item_price"] * row["quantity"])
    
    revenue += float(row["item_price"].replace("$","")) * row["quantity"]
    
revenue
    
```

```python
#5. How many orders were placed at this chipotle?
chipotle_orders = chipotle_data.shape[0]
chipotle_orders
```

```python
#6. How many items cost most than $10?
chipotle_prices = []
for index, row in chipotle_data.iterrows():
    
    chipotle_price = float(row["item_price"].replace("$", ""))
    chipotle_prices.append(chipotle_price)

chipotle_data['item_price'] = chipotle_prices
chipotle_data

price = len(chipotle_data[chipotle_data["item_price"]>10])
price
```
